```python
def shorten_categories(categories,cutoff):
    categorical_map={}

    for i in range(len(categories)):
        if categories.values[i]>= cutoff:
            categorical_map[categories.index[i]]=categories.index[i]
        else:
            categorical_map[categories.index[i]]="Other"
    return categorical_map
```

```python
country_map=shorten_categories(df['Country'].value_counts(),400)
country_map
```

```python
df['Country' ] = df['Country'].map(country_map)

df['Country'].value_counts()
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_experience(exp):
    if exp == "More than 50 years":
        return 50
    if exp == "Less than 1 year":
        return 0.5
    return float(exp)

df["YearsCodePro"] = df["YearsCodePro"].apply(clean_experience)
```

```python
df["YearsCodePro"].unique()
```

```python
df['EdLevel'].unique()
```

```python
def clean_education(edu):
    if 'Bachelorâ€™s degree' in edu:
        return 'Bachelors degree'
    elif 'Masterâ€™s degree' in edu:
        return 'Masters degree'
    elif 'Professional degree' in edu or 'other doctoral degree' in edu:
        return 'Post grad'
    return 'Less than a Bachelors'

df["EdLevel"]=df["EdLevel"].apply(clean_education)
```

```python
df['EdLevel'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_education=LabelEncoder()
df['EdLevel']=le_education.fit_transform(df['EdLevel'])
```

```python
le_country=LabelEncoder()
df['Country']=le_country.fit_transform(df['Country'])
df
```

MODEL TRAINING

1. LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

linear_reg = LinearRegression()
linear_reg.fit(X,Y.values)
```

```python
y_pred=linear_reg.predict(X)
```

```python
error=np.sqrt(mean_squared_error(Y,y_pred))
error
```

2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

decision_tree= DecisionTreeRegressor()
decision_tree.fit(X,Y.values)
```

```python
y_pred_dec=decision_tree.predict(X)
```

```python
error=np.sqrt(mean_squared_error(Y,y_pred_dec))
error
```

3. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

Random_forest=RandomForestRegressor()
Random_forest.fit(X,Y.values)
```

```python
y_pred_ran=Random_forest.predict(X)
```

```python
error=np.sqrt(mean_squared_error(Y,y_pred_ran))
error
```
